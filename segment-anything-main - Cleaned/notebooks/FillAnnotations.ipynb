```python
import pandas as pd

```

```python


df = pd.read_csv(r'C:\Users\chapi\Documents\STAGE\CODE\segment-anything-main\notebooks\SecondVersion_Annotations_Filled.csv')

for i in range(len(df)):
    regions = 0
    trash = 0
    row = df.iloc[i]

    if row['region_attributes'] == '{"Object":"Eel"}' or row['region_attributes'] == '{"Object":"Trash"}':
    

        filtered_df = df[df['filename'] == row['filename']]

        for j in range(len(filtered_df)):
            if filtered_df["region_attributes"].iloc[j] == '{"Object":"Eel"}':
                regions = 1
                break
        
        for k in range(len(filtered_df)):
            if filtered_df["region_attributes"].iloc[k] == '{"Object":"Trash"}':
                if trash == 2:
                    break
                trash = trash + 1
                regions = regions + 1
        
        df.loc[(df['filename'] == row['filename']),"file_attributes"] = '{"Object_Count":"'+str(regions)+'"}'


        print(filtered_df)
        

df.loc[df['file_attributes']=='{}','file_attributes'] = '{"Object_Count":"0"}' 

df.to_csv('SecondVersion_Annotations_Filledcc.csv', index=False,header=True)
```
